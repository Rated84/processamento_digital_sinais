1- Muitos computadores e calculadoras calculam a raiz quadrada de um número positivo 𝐴 através do seguinte algoritmo recursivo:

$$y[n] = \frac{1}{2}\left(y[n-1] + \frac{x[n]}{y[n-1]}\right)$$

em que 𝑥[𝑛] = 𝐴𝑢[𝑛], sendo 𝐴 o valor no qual deseja-se estimar a raiz quadrada e 𝑦[−1] e uma estimativa inicial para a raiz quadrada de 𝐴. O algoritmo é resolvido de forma recursiva para 𝑛 ≥ 0. Após uma quantidade satisfatórias de interações, 𝑦[𝑛] tenderá ao valor real de √ 𝐴. Implemente esse algoritmo em Python ou Matlab. Verifique seus resultados para os seguintes casos, em que 𝑁 é a quantidade de interações realizadas pelo algoritmo. Como estimativa inicial, utilize 𝑦[−1] = 𝐴/2.



```python
# a)
import math
import numpy as np
import matplotlib.pyplot as plt
A = 5
N = 10

y = A / 2
Y = [y]

i = 0;
erroA = []
for i in range(N):
    y = (y + A / y)/2
    Y.append(y)
    erroA.append(abs(y - math.sqrt(A)))


print("\n")
print(f"Resultado após {N} iterações: y = {y}\n")
print(f"Valor real da raiz de A: {math.sqrt(A)}\n")
print(f"Erro absoluto final: {erroA}\n")


plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(Y)
plt.title("Raiz Quadrada de {}".format(A))
plt.subplot(2, 1, 2)
plt.plot(erroA)
plt.title("Erro em Cada Iteração")
plt.tight_layout()

print(erroA)

plt.show()
```
